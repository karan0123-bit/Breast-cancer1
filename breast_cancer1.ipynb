# splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y= train_test_split(x1,y1,test_size=0.2,random_state=20)
```

# Import Support Vector Machine(SVM)

```python
from sklearn.svm import SVC
```

```python
svc_model=SVC()
```

# Now lets's train the model

```python
svc_model.fit(train_x,train_y)
```

# predict the data

```python
pred=svc_model.predict(test_x)
```

# The confusion matrix

```python
# Import metric libraries
from sklearn.metrics import classification_report,confusion_matrix
```

```python
import numpy as np 
np.where(pred!=test_y)
```

```python
cm=confusion_matrix(pred,test_y)
cm
```

```python
TN=cm[0,0]
TP=cm[1,1]
FN=cm[1,0]
FP=cm[0,1]
```
